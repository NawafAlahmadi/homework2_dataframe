# (1) Create script to pars json file into DataFrame.

# (2)remove punctuation

```python
import string
import re
```

```python

def remove_punctuation(text):

    punct_pattern = re.compile('[{}]'.format(re.escape(string.punctuation)))
    

    cleaned_text = punct_pattern.sub('', text)
    
    return cleaned_text


df_script_jf['headline'] = df_script_jf['headline'].apply(remove_punctuation)

df_script_jf['short_description'] = df_script_jf['short_description'].apply(remove_punctuation)
df_script_jf
```

# (3)tokenize 

```python
import re
def tokenize(text):
    tokens = re.split('\W+' , text )
    return tokens

df_script_jf['headline_tokenize'] = df_script_jf['headline'].apply(lambda x: tokenize(x.lower()))
df_script_jf['short_description_tokenize'] = df_script_jf['short_description'].apply(lambda x: tokenize(x.lower()))
df_script_jf
```

# (4) remove stopwords

```python
import nltk 

stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(tokenized_text):
    text = [word for word in tokenized_text if word not in stopword]
    return text 
df_script_jf['headline_nostop'] = df_script_jf['headline_tokenize'].apply(lambda x: remove_stopwords(x))
df_script_jf['short_description_nostop'] = df_script_jf['short_description_tokenize'].apply(lambda x: remove_stopwords(x))
df_script_jf
```
